```python
import cv2
import common
import pylab
import cv2
import pytesseract
import matplotlib.pyplot as plt
#Para windows
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
placa1 = []
image1 = cv2.imread('Imagenes/ej01-auto.jpg')
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB));
```

```python
count = 0
while count <= 2:
    gray1 = cv2.blur(gray1, (3,3))
    canny1 = cv2.Canny(gray1, 130,200)
    canny1 = cv2.dilate(canny1, None, iterations=1)
    count += 1
plt.imshow(canny1, cmap='gray')
```

```python
# Deteccion de contornos con OpnenCV
contorno1, _ = cv2.findContours(canny1, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for c in contorno1:
    area = cv2.contourArea(c)
    x, y, w, h = cv2.boundingRect(c)
    epsilon = 0.9 * cv2.arcLength(c, True)
    aprox = cv2.approxPolyDP(c, epsilon, True)
    if area >= 5000:
        print('area',area)
        print('x:', x)
        print('y:', y)
        print('w:', w)
        print('h:', h)
        relAspec = float(w)/h
        print('Relacion de aspecto:', relAspec)
        cv2.drawContours(image1, [aprox], 0, (0,255,0), 3)
        if relAspec > 1:
            placa1 = gray1[y:y+h, x:x+w]
            text = pytesseract.image_to_string(placa1, config='--psm 11'.strip())
            print('Placa: ', text)
            cv2.rectangle(image1, (x, y), (x+w, y+h), (0,255,0), 3)
            cv2.putText(image1, text, (x-20, y-20), 1, 2.2, (0,255,0), 3)
```

```python
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB));
```
