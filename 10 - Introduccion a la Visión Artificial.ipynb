```python
# Importar paquetes
import matplotlib.pyplot as plt
import numpy as np
import imageio
from skimage import color
```

```python
# Lectura de la imagen
ruta = r'C:\Users\medin\OneDrive\Documentos\Ingenier√≠a en Sistemas\Primer Semestre 2,025\inteligencia artificial\gray-wolf.jpg'
imgIn = imageio.imread(ruta)
```

```python
#Imprimir el contenido de las capas de color
print('Pixel en [0, 0, 0]', imgIn[0, 0, 0])
print('Pixel en [0, 0, 1]', imgIn[0, 0, 1])
print('Pixel en [0, 0, 2]', imgIn[0, 0, 2])
print('Pixel en [0, 0, :]', imgIn[0, 0, :])
```

```python
print('Dimensiones:', imgIn.shape)
print('Tipo de dato:', imgIn.dtype)
```

```python
# Rango de valores contenidos en la matriz (por color)
print('Valor maximo:', imgIn.max())
print('Valor miminimo:', imgIn.min())
```

# Visualizacion de la imagen

```python
plt.imshow(imgIn);
```

## Transformacion a escala de grises

```python
imgGray = color.rgb2gray(imgIn)
print('Dimensiones en escala de grises:', imgGray.shape)
print('Tipo de dato contenido:', imgGray.dtype)
```

```python
# Imagen en escala de grises?
plt.imshow(imgGray);
```

```python
plt.imshow(imgGray, cmap='gray');
```

```python
# Seccion de imagen
imgSeccion = imgGray[110:440, 215:525]
plt.imshow(imgSeccion, cmap='gray');
```

```python
# Cambio de valores en la matriz de color
imgModificada = imgGray.copy()
imgModificada[imgModificada < 0.2] = 0.0
plt.imshow(imgModificada, cmap='gray');
```

```python
# Distribucion de valores de los pixeles
plt.hist(imgGray.flatten());
```
