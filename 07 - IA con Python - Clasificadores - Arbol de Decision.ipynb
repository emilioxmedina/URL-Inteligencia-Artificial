# Inteligencia Artificial con Python: Clasificadores

## Arbol de decision

```python
# importar bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# importar el separador de muestras para entrenamiento y pruebas
from sklearn.model_selection import train_test_split
# importar el clasificador: arbol de decision
from sklearn.tree import DecisionTreeClassifier, plot_tree
# importar metricas para medir la eficiencia del modelo
from sklearn import metrics
```

```python
# Cargar el dataset
data = sns.load_dataset('iris')
```

```python
train, test = train_test_split(data, test_size = 0.4, stratify = data['species'], random_state = 42)
```

```python
# Separacion de la etiqueta de clase <<Y>> y atributos<<X>>
# Conjunto de entrenamiento
x_train = train[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y_train = train['species']
# Conjunto de pruebas
x_test = test[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y_test = test['species']
```

```python
# Arbol de decision
mod_dt = DecisionTreeClassifier(max_depth = 3, random_state = 1)
# Ajustar/entrenar el modelo
mod_dt.fit(x_train, y_train)
# Predecir resultados sobre el conjunto de pruebas
prediccion = mod_dt.predict(x_test)
```

```python
# Importancia de cada predictor

mod_dt.feature_importances_
```

```python
# Etiquetas de datos
fn = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
cn = ['setosa', 'versicolor', 'virginica']
```

```python
# Visualizar las reglas de clasificacion
plt.figure(figsize = (10, 8))
plot_tree(mod_dt, feature_names=fn, class_names=cn, filled=True);
```

```python
eficiencia = metrics.accuracy_score(y_test, prediccion)
eficiencia
```
