```python
import math
import numpy as np
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
from scipy import stats
```

```python
data = sns.load_dataset('iris')
print(data.head(10))
```

```python
data['species'].unique()
```

```python
x = data['petal_length']
y = data['petal_width']
```

```python
x = sm.add_constant(x)
```

```python
resultado = sm.OLS(y, x).fit()
resultado.summary()
```

```python
resultado.params
```

```python
print(resultado.rsquared)
print(math.sqrt(resultado.rsquared))
```
